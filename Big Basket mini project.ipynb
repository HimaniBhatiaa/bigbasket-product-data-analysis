#Step 8: Find out the outliers from the dataset according to the columns and fill them with the mean.

###Finding outliers in the three columns i.e, index,sale_price, and market_price

```python
Q1 = BigBasketData['index'].quantile(0.25)
Q3 = BigBasketData['index'].quantile(0.75)
IQR = Q3-Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = BigBasketData[(BigBasketData['index'] < lower_limit) | (BigBasketData['index'] > upper_limit)]
print(outliers)
```

```python
Q1 = BigBasketData['sale_price'].quantile(0.25)
Q3 = BigBasketData['sale_price'].quantile(0.75)
IQR = Q3-Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = BigBasketData[(BigBasketData['sale_price'] < lower_limit) | (BigBasketData['sale_price'] > upper_limit)]
print(outliers)
```

```python
Q1 = BigBasketData['market_price'].quantile(0.25)
Q3 = BigBasketData['market_price'].quantile(0.75)
IQR = Q3-Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = BigBasketData[(BigBasketData['market_price'] < lower_limit) | (BigBasketData['market_price'] > upper_limit)]
print(outliers)
```

###Now finding out their mean and filling them with their mean

```python
BigBasketData["index"].mean()
```

```python
BigBasketData["sale_price"].mean()
```

```python
BigBasketData["market_price"].mean()
```

```python
BigBasketData["index"]=BigBasketData["index"].fillna(BigBasketData["index"].mean())
```

```python
BigBasketData["sale_price"]=BigBasketData["sale_price"].fillna(BigBasketData["sale_price"].mean())
```

```python
BigBasketData["market_price"]=BigBasketData["market_price"].fillna(BigBasketData["market_price"].mean())
```

#Step 9: Create Plots or visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##1. HIST PLOT (To understand price range and spread)

```python
sns.histplot(BigBasketData["market_price"], kde=True)
plt.title("Distribution of market price")
plt.xlabel("market Price")
plt.ylabel("Count")
plt.show()
```

##Boxplot for Outlier Visualization


```python
sns.boxplot(x=BigBasketData["market_price"])
plt.title("Boxplot for market price")
plt.show()
```

##Scatter plot for comparing sale price and market price to visalize discounts



```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='market_price', y='sale_price', data=BigBasketData)
plt.title("market price vs sale price")
plt.xlabel("market_price")
plt.ylabel("sale_price")
plt.show()
```

##Barplot to see category-wise average sale price to see which product is costliest on average


```python
sns.barplot(x='category', y='market_price', data=BigBasketData)
plt.title("average sale price per category")
plt.xticks(rotation=45)
plt.show()
```

##Correlation heatmap

```python
sns.heatmap(BigBasketData[["sale_price","market_price"]].corr(),annot=True)
plt.title("correlation heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(BigBasketData[['index', 'rating', 'Discount']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Index, Rating, and Discount')
plt.show()
```
